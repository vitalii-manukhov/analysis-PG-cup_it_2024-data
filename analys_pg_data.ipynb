### Импорт библиотек

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_corr_matrix(corr_matrix):
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix,
                annot=True,
                cmap="coolwarm")
    plt.show()
```

##### Период времени, охваченный датами

```python
# Даты в формате ISO 8601
pos_general_copy = pos_general["last_sync_date"]
pos_general_copy = pd.to_datetime(pos_general_copy)

min_date = pos_general_copy.min()
max_date = pos_general_copy.max()

period = max_date - min_date
print(period)
```
